```python
from math import tanh,cosh
from numpy import linspace
from pylab import plot,show,ylim,xlabel,ylabel

# Constantes
Tmax = 2.0
points = 1000
accuracy = 1e-6

# Configurando listas para graficar
y = []
temp = linspace(0.01,Tmax,points)

# Bucle Temperatura
for T in temp:
    m1 = 1.0
    error = 1.0
    
    # Bucle hasta que el error sea lo suficiente pequeño
    while error>accuracy:
        m1,m2 = tanh(m1/T),m1
        error = abs((m1-m2)/(1-T*cosh(m2/T)**2))
    y.append(m1)

# Hacer la grafica
plot(temp,y)
ylim(-0.1,1.1)
xlabel("Temperatura")
ylabel("Magnetizacion")
show()
```
